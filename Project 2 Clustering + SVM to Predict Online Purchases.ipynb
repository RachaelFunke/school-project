### Outliers handling

```python
from scipy.stats import zscore

# Columns to check for outliers
columns_to_check = ['Administrative_Duration', 'Informational_Duration', 'PageValues']

# Calculate Z-scores
z_scores = df[columns_to_check].apply(zscore)

# Define a threshold for identifying outliers
threshold = 3

# Replace data points where the absolute value of the Z-score is greater than the threshold
for column in columns_to_check:
    # Find outliers
    outliers = np.abs(z_scores[column]) > threshold
    
    # Handle outliers, for example, by capping
    capped_values = np.where(outliers, z_scores[column].apply(lambda x: threshold if x > 0 else -threshold), z_scores[column])
    
    # Replace data in the original DataFrame
    df[column] = capped_values
```

```python
# Now let's plot the boxplots for these columns after handling outliers
import matplotlib.pyplot as plt

# Plot
fig, axes = plt.subplots(nrows=len(columns_to_check), ncols=1, figsize=(10, 15))
for i, column in enumerate(columns_to_check):
    df[column].plot(kind='box', ax=axes[i], title=column)
    axes[i].set_ylabel('Z-scores')

plt.tight_layout()
plt.show()
```

__Observation__

After capping the outliers in the dataset using the Z-score method with a threshold of ±3, the boxplots for the columns 'Administrative_Duration', 'Informational_Duration', 'ProductRelated_Duration', and 'PageValues' display a reduction in the range of extreme values, indicating a more compact and normalized distribution that is less influenced by extreme deviations. This preprocessing should lead to more robust clustering results, as the algorithm's performance is less likely to be adversely affected by the presence of outliers.

### Elbow Method

```python
# Calculate distortions for a range of number of clusters
inertia = []
for i in range(1, 12):
    kmeans = KMeans(n_clusters=i, random_state=0).fit(df_transformed)
    inertia.append(kmeans.inertia_)

# Plot
plt.figure(figsize=(8, 4))
plt.plot(range(1, 12), inertia, marker='o')
plt.title('Elbow Method For Optimal k')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

__observation__

The elbow graph presented displays a gradual decrease in inertia, with no distinct "elbow" point where the rate of decrease sharply changes. However, there's a slight inflection around 4 clusters, suggesting that increasing the number of clusters beyond 4 yields lesser benefits in reducing inertia, thus 4 could be considered as a potential choice for the number of clusters for the dataset.

```python
silhouette_scores = []
range_n_clusters = list(range(2, 12))

for n_clusters in range_n_clusters:
    clusterer = KMeans(n_clusters=n_clusters, random_state=10)
    cluster_labels = clusterer.fit_predict(df_transformed)
    silhouette_avg = silhouette_score(df_transformed, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For n_clusters = {n_clusters}, The average silhouette_score is : {silhouette_avg}")
```

```python
# Plot
plt.plot(range_n_clusters, silhouette_scores, marker='o')
plt.title('Silhouette analysis')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```

__Observation__

The silhouette analysis graph shows a peak at 6 clusters, indicating that a configuration of six clusters maximizes the average silhouette score, thus suggesting that six clusters may provide a reasonable balance between cohesion within clusters and separation between different clusters in the dataset

### Constructing SVM models

I've created two SVM models, each with a different subset of explanatory variables. The first focuses on user engagement, the second on traffic and regional features

```python
# Define feature sets for two models
features_model_1 = ['Administrative_Duration', 'Informational_Duration', 'ProductRelated', 'ExitRates', 'PageValues']
features_model_2 = ['TrafficType_2', 'Region_1', 'Month_Nov']

# Separate the feature sets and target variable
X1 = concatenated_df[features_model_1]
X2 = concatenated_df[features_model_2]
y = concatenated_df['Revenue']  # or 'V_Revenue' depending on the chosen response variable

# Create training and testing sets for each model
X1_train, X1_test, y_train, y_test = train_test_split(X1, y, test_size=0.2, random_state=42)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y, test_size=0.2, random_state=42)

# Initialize SVM models with a pipeline that includes scaling
svm_model_1 = make_pipeline(StandardScaler(), SVC(kernel='linear'))
svm_model_2 = make_pipeline(StandardScaler(), SVC(kernel='linear'))

# Fit the models
svm_model_1.fit(X1_train, y_train)
svm_model_2.fit(X2_train, y_train)
```


### Evaluate the models

```python
# Evaluate Model 1
y_pred_1 = svm_model_1.predict(X1_test)
report_1 = classification_report(y_test, y_pred_1, output_dict=True)
accuracy_1 = accuracy_score(y_test, y_pred_1)

# Evaluate Model 2
y_pred_2 = svm_model_2.predict(X2_test)
report_2 = classification_report(y2_test, y_pred_2, output_dict=True)
accuracy_2 = accuracy_score(y2_test, y_pred_2)

# Convert classification reports to DataFrames
df_report_1 = pd.DataFrame(report_1).transpose()
df_report_2 = pd.DataFrame(report_2).transpose()

# Add accuracy to the DataFrames
df_report_1['accuracy'] = accuracy_1
df_report_2['accuracy'] = accuracy_2

# Combine the reports into a single DataFrame for easier comparison
# Note: The indices here are set for easier readability in the comparison DataFrame
comparison_df = pd.concat([df_report_1.add_prefix('Model 1 - '), df_report_2.add_prefix('Model 2 - ')], axis=1)

comparison_df
```

The evaluation results state that Model 1 outperforms Model 2 significantly across all key metrics, including precision, recall, and F1-score, both for the classification of the '0' label and when aggregated (macro and weighted averages). Notably, Model 1 demonstrates perfect precision for the '1' label, whereas Model 2 completely fails to identify any '1' labels, suggesting that Model 1 has a superior balance of recall and precision and is much better suited for making balanced predictions across different classes.

### Cross Validation

### ROC Curve

```python
from sklearn.metrics import roc_curve, confusion_matrix, roc_auc_score
import matplotlib.pyplot as plt

# Calculate the false positive rate, true positive rate and thresholds for both models
fpr1, tpr1, thresholds1 = roc_curve(y_test, y_pred_1)
fpr2, tpr2, thresholds2 = roc_curve(y_test, y_pred_2)

# Calculate the AUC (Area under the ROC Curve) for both models
roc_auc1 = roc_auc_score(y_test, y_pred_1)
roc_auc2 = roc_auc_score(y_test, y_pred_2)

# Create Confusion Matrix for both models
conf_matrix1 = confusion_matrix(y_test, y_pred_1)
conf_matrix2 = confusion_matrix(y_test, y_pred_2)
```

```python
# Plot ROC Curve for Model 1
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(fpr1, tpr1, color='blue', label=f'Model 1 AUC = {roc_auc1:.2f}')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Model 1')
plt.legend(loc="lower right")

# Plot ROC Curve for Model 2
plt.subplot(1, 2, 2)
plt.plot(fpr2, tpr2, color='green', label=f'Model 2 AUC = {roc_auc2:.2f}')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Model 2')
plt.legend(loc="lower right")

plt.tight_layout()
plt.show()
```

__Observation__

The ROC curves shows that Model 1 significantly outperforms Model 2, as indicated by the area under the curve (AUC) where Model 1's AUC is close to 1, reflecting a high true positive rate across thresholds with a low false positive rate, while Model 2's AUC of 0.50 is equivalent to random guessing, indicating no discriminative power between the positive and negative classes.
